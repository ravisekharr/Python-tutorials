**Question 2**

```python
#A Read dataset
df_sales=pd.read_excel("/content/Cust_Sales.xlsx")
df_sales=df_sales[['Customer', 'Store', 'Category', 'Birth Date', '1-Jan', '1-Feb',
       '1-Mar', '1-Apr', '1-May', '1-Jun', '1-Jul', '1-Aug',
       '1-Sep', '1-0ct', '1-Nov', '1-Dec']]
df_sales
```

```python
#B: Fill NA in birth date column using mode
df_sales['Birth Date']=df_sales['Birth Date'].fillna(df_sales['Birth Date'].mode()[0])
df_sales
```

```python
#C: Delete records for X29 store
df_sales.drop(df_sales[df_sales['Store']=='X29'].index,inplace=True)
df_sales
```

```python
#D: Aggregate data at store category level
agg_dict={}
aggcols=[x for x in df_sales.columns.values if '1' in x]
for x in aggcols:
  agg_dict[x]='mean'
print('Mean monthly sales')
print(df_sales.groupby(['Store','Category']).agg(agg_dict))
```
